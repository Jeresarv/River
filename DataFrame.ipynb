```python
import pandas as pd

# Crear un diccionario con los datos
datos = {
    "Alumno": ["Juan", "María", "Pedro"],
    "Curso": ["6°A", "6°B", "6°A"],
    "Nota": [8, 9, 7]
}

# Crear el DataFrame
df = pd.DataFrame(datos)

# Mostrar el DataFrame completo
print("DataFrame completo:\n")
print(df)

# Mostrar solo la columna de cursos
print("\nColumna de cursos:\n")
print(df["Curso"])

import pandas as pd

# Crear un diccionario con los productos
datos = {
    "Producto": ["Pan", "Leche", "Arroz", "Aceite"],
    "Precio": [500, 1200, 800, 2500]
}

# Crear el DataFrame
df = pd.DataFrame(datos)

# Agregar la columna IVA (21% del precio)
df["IVA"] = df["Precio"] * 0.21

# Mostrar el DataFrame
print(df)

import pandas as pd

# Crear un diccionario con ciudades y población
datos = {
    "Ciudad": ["Buenos Aires", "Córdoba", "Rosario", "Mar del Plata", "Salta"],
    "Poblacion": [3000000, 1400000, 1200000, 600000, 70000]
}

# Crear el DataFrame
df = pd.DataFrame(datos)

# Mostrar el DataFrame completo
print("DataFrame completo:\n")
print(df)

# Filtrar las ciudades con más de 100.000 habitantes
print("\nCiudades con más de 100.000 habitantes:\n")
print(df[df["Poblacion"] > 100000])

import pandas as pd

# Crear el DataFrame con materias y notas
datos = {
    "Materia": ["Matemática", "Lengua", "Historia", "Geografía", "Inglés"],
    "Nota": [8, 5, 7, 4, 9]
}

df = pd.DataFrame(datos)

# Mostrar el DataFrame completo
print("Notas del alumno:\n")
print(df)

# Filtrar materias aprobadas (nota >= 6)
aprobadas = df[df["Nota"] >= 6]

print("\nMaterias aprobadas:\n")
print(aprobadas)

```
