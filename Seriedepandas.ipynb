```python
import pandas as pd
cursos = pd.Series(
    [28, 30, 25, 32],  # cantidad de alumnos
    index=["6°A", "6°B", "5°A", "5°B"]  # nombres de cursos
)
# Mostrar la Serie completa
print("Serie de cursos y alumnos:")
print(cursos)

# Mostrar el número de alumnos del primer curso
print("\nNúmero de alumnos del primer curso:")
print(cursos.iloc[0])

import pandas as pd

# Crear la Serie con materias y notas
notas = pd.Series(
    [8, 7, 9, 6, 10],  # notas del alumno
    index=["Matemática", "Lengua", "Historia", "Ciencias", "Inglés"]  # materias
)

# Mostrar la Serie completa
print("Notas del alumno en cada materia:")
print(notas)

# Mostrar la nota de Matemática
print("\nNota en Matemática:")
print(notas["Matemática"])

import pandas as pd

# Primera Serie con productos y precios
productos1 = pd.Series(
    [1200, 800, 500],
    index=["Notebook", "Mouse", "Teclado"]
)

# Segunda Serie con productos y precios distintos
productos2 = pd.Series(
    [1000, 600, 300],
    index=["Notebook", "Monitor", "Impresora"]
)

# Mostrar las dos Series
print("Productos 1:")
print(productos1)

print("\nProductos 2:")
print(productos2)

# Sumar las Series
suma = productos1 + productos2

print("\nSuma de las Series:")
print(suma)

import pandas as pd

# Crear la Serie con temperaturas de 7 días
temperaturas = pd.Series(
    [22, 25, 19, 28, 30, 24, 27],
    index=["Lunes", "Martes", "Miércoles", "Jueves", "Viernes", "Sábado", "Domingo"]
)

# Mostrar la Serie completa
print("Temperaturas de la semana:")
print(temperaturas)

# Mostrar la temperatura máxima
print("\nTemperatura máxima de la semana:")
print(temperaturas.max())

import pandas as pd

# Crear la Serie con profesores y antigüedad
profesores = pd.Series(
    [12, 8, 15],  # años de antigüedad
    index=["Profesor Gómez", "Profesora López", "Profesor Martínez"]
)

# Mostrar la Serie completa
print("Antigüedad de los profesores:")
print(profesores)

# Mostrar el promedio de antigüedad
print("\nPromedio de antigüedad:")
print(profesores.mean())
```
